  ## Introduction à l'exploration univariée graphique
 L'exploration univariée graphique s'intéresse à la visualization des données par rapport à une seule variable à la fois. Plusieurs techniques de visualization peuvent être utilisées pour l'analyse exploratoire univariée telles que: les histogrammes, les boites à moustâches, les diagrammes en bâtons, les nuages de points, etc. Le choix d'une technique ou d'une autre dépend essentiellement du type de la variable à l'étude.

 ### Étapes d'analyse univariée graphique en Python

Différentes étapes sont nécessaires pour la visualization univariée:
1. Importer les modules Python necessaires
2. Charger les données et les sauvegarder dans un DataFrame
3. Visualizer les données à l'aide d'un graphique




 ## Importer les modules Python pour l'analyse de données

Pour l'analyse univariéé graphique, le module *pandas* est essentielle. Il dispose des *dataframes* qui représentent des structure de données puissantes pour repésenter les données en tableaux relationnels avec des lignes et des colonnes. L'importation du module *pandas* se fait comme suit:

```python
import pandas as pd
```

*Scikit-learn* est un autre module important pour l'analyse des données et surtout pour l'apprentissage automatique, il dispose d'un ensemble d'algorithmes et de données qui peuvent être utilisées pour toute sorte d'analyse potentielle. La ligne de code suivante permet d'importer des données sur le cancer du sein disponible dans le module *Scikit-learn*:

```python
from sklearn.datasets import load_breast_cancer
```

En plus des deux modules mentionnés ci-haut s'ajoute les modules de visualization, les plus polulaires sont *maplotlib* et *seaborn*. *Seaborn* est sur *matplotlib*, il fournit une interface de haut niveau pour dessiner des graphiques statistiques attrayants et informatifs.
Exécutez la commande ci-dessus pour importer le module de *seaborn*:

```python
import seaborn as sns
```

 ## Charger les données et les sauvegarder dans un DataFrame

Afin de pouvoir effectuer une analyse univariée sur *Python*, il est necéssaire de charger les données et les sauvegarder dans un *DataFrame*. Par exemple pour les données du cancer du sein de Scikit-learn, on peut charger les données dans un *DataFrame* en exécutant la cellule suivante:

```python
cancerData = load_breast_cancer()
cancer = pd.DataFrame(cancerData.data, columns=cancerData.feature_names)
```

## Visualizer les données à l'aide d'un graphique

Le module *seaborn* dispose d'un panoplie de graphiques pour la visualization univariée. Le choix d'un graphique parmi d'autres dépends du type de données à visualizer (quantitatives ou qualitatives). Par exemple, pour les données quantitatives, un histogramme ainsi que les boîtes à moustâches permettent d'illustrer la distribution des données par rapport à une variable précise. Les boîtes à moustâches permettent, de plus, à visualizer les valeurs abérrantes.
Exécutez la cellule suivante pour visualizer un histogramme représentant la distribution de la variable *mean radius* dans les données du cancer de sein. N'oubliez pas d'exécuter toutes les lignes de code ci-dessus avant.


```python
sns.histplot(cancer['mean radius'])
```

 ## Exercice

```python
sns.boxplot(data=cancer['mean radius'])
```
