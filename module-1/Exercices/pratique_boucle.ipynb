# Solutionnaires

```python
import json
fici = open("contacts.json")
 
reponse = json.load(fici)
#print(reponse)
data = reponse["contacts"]

for rec in data:
    for k,v in rec.items():
        print(k,":",v)
    print("=" * 50)
```
